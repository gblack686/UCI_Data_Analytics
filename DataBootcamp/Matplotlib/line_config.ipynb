```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0,10,.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
plt.hlines(0,0,10,alpha = .25)
```

```python
sine_handle, = plt.plot(x_axis,sin,marker = 'o', color = 'blue', label = 'sin')
cos_handle, = plt.plot(x_axis,cos,marker = 'v', color = 'red', label = 'cos')
```

```python
plt.legend(loc = "upper right")
```
