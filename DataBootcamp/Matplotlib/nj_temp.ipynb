```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
points
```

```python
plt.plot(x_axis_data,points)
```

```python
points_c = [(x-32)*.56 for x in points]
points_c
```

```python
plt.plot(x_axis_data,points_c)
```

```python
f_handle, = plt.plot(x_axis_data,points,marker = 'o', color = 'blue', label = 'Fahrenheit')
c_handle, = plt.plot(x_axis_data,points_c,marker = 'v', color = 'red', label = 'Celcius')
```

```python
plt.legend(loc = "best")
```

```python
fahrenheit, = plt.plot(x_axis_data, points, marker ='+', color = "blue", linewidth =1, label = "Fahrenheit")
celcius, = plt.plot(x_axis_data, points_c, marker ='o', color = "red", linewidth =1, label = "Celcius")
```

```python
plt.xlabel("Months")
plt.ylabel("Degrees")
```
